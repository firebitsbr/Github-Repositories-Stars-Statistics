# Github Repositories Stargazers Statistics

This notebook gets current stargazers for some repositories and then it views the daily stargazers report within a specific range.

Repositories with 50000 stars or more than should be excluded, because Github API won't return all that data in this case.

```python
# Imports

import numpy as np
from github import Github
from IPython.display import clear_output
from datetime import date, datetime, timedelta
from dateutil import parser
import matplotlib.pyplot as plt


%matplotlib inline
```

```python
# Settings

###############################Inputs###########################################

username = ""              # Enter username here.. (Required)
password = ""         # Enter password here.. (Required)

# Enter some valid repositories name: username/rep_name .. (At least one is required)
repositories_names = ["mbadry1/DeepLearning.ai-Summary", "mbadry1/Top-Deep-Learning", "mbadry1/Trending-Deep-Learning"]          
# start_date = "2017-12-12 07:26:36"                 # Start date range.. (Optional)
# end_date = "2017-12-14 07:26:36"                   # end date range.. (Optional)
start_date = None                 # Start date range.. (Optional)
end_date = None                   # end date range.. (Optional)

###############################Inputs###########################################


all_reps = []
reps_stars_counts = {}
reps_stars_dates = {}
if start_date is not None:
    start_date = parser.parse(start_date)
if end_date is not None:
    end_date = parser.parse(end_date)
if (start_date is not None) and (end_date is not None) and (start_date > end_date):
    start_date, end_date = end_date, start_date
```

```python
# Functions
```

```python
# Github object
g = Github(username, password)

# Getting valid repositories
all_reps = []
for name in repositories_names:
    rep_found = True
    try:
        r = g.get_repo(name, lazy=False)
    except:
        rep_found = False
    
    if rep_found:
        all_reps.append(r)
        print(("'{}'repository loaded.").format(name))
        
if not all_reps:
    print("Warning: At least one valid repository is required!")
```

```python
# Reading the stargazers into lists
reps_stars_counts = {}
reps_stars_dates = {}

for rep in all_reps:
    print("Working with:", rep.full_name)
    rep = g.get_repo(rep.full_name)       # Fetching it again, Can be commented if its one time running.
    
    reps_stars_counts[rep.full_name] = []
    reps_stars_dates[rep.full_name] = []
    
    current_day_count = 0
    current_day = None
    for stargazer in rep.get_stargazers_with_dates():
        d = stargazer.starred_at       
        
        if (start_date is not None) and (d < start_date):
            continue
        if (end_date is not None) and (d > end_date):
            break   # End. because the stars are sorted by date
        
        d = d.date()
        # print(d)
        if current_day != d:
            if (current_day is not None):
                reps_stars_counts[rep.full_name].append(current_day_count)
                reps_stars_dates[rep.full_name].append(current_day)
            current_day = d
            current_day_count = 1
        else:
            current_day_count += 1

    # Finalizing
    if (rep and current_day is not None):
        reps_stars_counts[rep.full_name].append(current_day_count)
        reps_stars_dates[rep.full_name].append(current_day)

print("Done.")
```

```python
# Refining time series

# First getting min and max dates
min_date = date.max
max_date = date.min
for k in reps_stars_dates:
    if reps_stars_dates[k][0] < min_date:
        min_date = reps_stars_dates[k][0]
    if reps_stars_dates[k][len(reps_stars_dates[k])-1] > max_date:
        max_date = reps_stars_dates[k][len(reps_stars_dates[k])-1]
        
        
# Adding missing dates day by day in the range min_date:max_date
for k in reps_stars_dates:
    i = 0
    
    # Inserting min and max dates
    if (reps_stars_dates[k][0] != min_date):
        reps_stars_dates[k].insert(0, min_date)
        reps_stars_counts[k].insert(0, 0)
    if (reps_stars_dates[k][len(reps_stars_dates[k])-1] != max_date):
        reps_stars_dates[k].append(max_date)
        reps_stars_counts[k].append(0)
    
    i = 0
    while True:
        if i+1 == len(reps_stars_dates[k]):
            break
        d = reps_stars_dates[k][i]
        next_d = reps_stars_dates[k][i+1]
        d_plus_day = d + timedelta(days=1)
        
        if next_d != d_plus_day:
            reps_stars_dates[k].insert(i+1, d_plus_day)
            reps_stars_counts[k].insert(i+1, 0)
        i += 1
```

```python
# Plotting charts


fig = plt.figure(figsize=(15, 12))

handles = []
for k in reps_stars_counts:
    l, =plt.plot(reps_stars_dates[k], reps_stars_counts[k], label=k)
    handles.append(l)
plt.legend(handles=handles, loc='best')
    
```
